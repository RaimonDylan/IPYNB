## Calculez le RMS de ce signal et retournez le en dB SPL.

```python
import numpy as np      # Importer le module numpy
np.random.seed(0) 
x = np.random.random(1000)
print(x[:3]) 
meh = np.sqrt((x**2).mean())
## RMS du tableau
db = 20.0*np.log10(meh*20e-6)
print(db)
```
